```python
import json
from pathlib import Path
import os
import pandas as pd
```

```python
file_path=os.path.join(Path().resolve().parent,'reports')
print(file_path)
```

```python
m1=json.load(open(os.path.join(file_path,'metrics_1.json'),'r'))
m2=json.load(open(os.path.join(file_path,'metrics_2.json'),'r'))
m3=json.load(open(os.path.join(file_path,'metrics_3.json'),'r'))
```

```python
rows = []
for entry in m1['full_result']:
    fold = f"Fold{entry['fold']}"
    for model in entry['result']:
        row = {
            'index': fold,
            'sub_index': model['name'],
            'f1_score': model['metrics']['f1_score'],
            'roc_auc_score': model['metrics']['roc_auc_score'],
            "concordance_index_score": model['metrics']['concordance_index_score'],
            "race_score": model['metrics']['race_score']
        }
        rows.append(row)


df1 = pd.DataFrame(rows).set_index(['index', 'sub_index'])

df1
```
